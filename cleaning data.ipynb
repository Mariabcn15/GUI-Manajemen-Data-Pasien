```python
import pandas as pd
import numpy as np

# === Contoh: ganti dengan pembacaan file asli kamu ===
# df = pd.read_csv("data.csv")

df = cleaned_data

# === 1. Normalisasi teks & samakan sinonim ===
def clean_food(val):
    if pd.isna(val) or str(val).strip() == '':
        return np.nan
    val = val.strip().lower()
    mapping_synonym = {
        'alergi udang': 'udang',
        'udang ': 'udang',
        'seafood': 'seafood',
        'baik': np.nan,
        'tidak ada': np.nan,
        'tidak ada ': np.nan,
        'tidak_ada': np.nan
    }
    return mapping_synonym.get(val, val)

df['alergi_makanan'] = df['alergi_makanan'].apply(clean_food)

# === 2. Mapping kategori ke angka ===
# Ambil kategori unik yang valid (tidak termasuk NaN)
unique_cats = sorted(df['alergi_makanan'].dropna().unique())

# Buat mapping: mulai dari 1, NaN akan diberi 0
cat_to_num = {cat: idx+1 for idx, cat in enumerate(unique_cats)}
cat_to_num[np.nan] = 0  # kosong → 0

# === 3. Ubah kolom aslinya menjadi numerik ===
df['alergi_makanan'] = df['alergi_makanan'].map(cat_to_num).fillna(0).astype(int)

# === 4. Lihat hasil ===
print("\nMapping kategori ke angka:", cat_to_num)
print(cleaned_data['alergi_makanan'].value_counts(dropna=True).head(20))
```

```python
# Tampilkan 30 baris teratas yang berisi alergi_obat
print(cleaned_data[cleaned_data['alergi_obat'].notna()]['alergi_obat'].head(70))

# Hitung berapa baris terisi
print("\nJumlah baris terisi:", cleaned_data['alergi_obat'].notna().sum())

# Lihat beberapa nilai unik terbanyak (top 50)
print(cleaned_data['alergi_obat'].value_counts(dropna=True).head(70))
```

```python
import pandas as pd
import re

df = cleaned_data

canonical_map = {
    'asam mefemanat': 'asam mefenamat',
    'penicilin': 'penisilin',
    'ampicilin': 'ampicillin',
    'amoxicilin': 'amoxicillin',
    'amoxcilin': 'amoxicillin',
    'amoxilin': 'amoxicillin',
    'amoxsilin': 'amoxicillin',
    'cefadroxsil': 'cefadroxil',
    'ketokonazole': 'ketoconazole',
    'cefixim': 'cefixime',
    'cipro': 'ciprofloxacin',
    'inj. cefrtriaxon': 'ceftriaxone',
    'cephalosporin jenis cefixime': 'cefixime',
    'methylprednisolon': 'methylprednisolone'
}

# Daftar obat standar (whitelist)
obat_standar = {
    'asam mefenamat', 'cefadroxil', 'lansoprazole', 'penisilin', 'kotrimoxazole',
    'antalgin', 'mylanta', 'ampicillin', 'amoxicillin', 'bodrexin', 'contrexin',
    'ketoconazole', 'eperison', 'cetirizine', 'metroklopramid hcl', 'ceftriaxone',
    'ketorolac', 'piroxicam', 'metampiron', 'cefixime', 'meloxicam', 'ciprofloxacin',
    'sulfa', 'tetra', 'eritromicin', 'methylprednisolone', 'paracetamol', 'piracetam',
    'natrium diclofenac', 'konidin', 'komix', 'salbutamol', 'avamys', 'dexa',
    'ambroxol', 'mefinal', 'metronidazole', 'ibuprofen'
}

# Kata-kata non-obat yang harus dibuang
stopwords_non_obat = {
    'tidak ada', 'baik', 'sejenis', 'apabila', 'reaksi alergi', 'ruam makulopapular',
    'injeksi', 'inj', 'jenis', 'gatal seluruh tubuh', 'apabila minum sakit perut'
}

def clean_alergi_obat_strict(text):
    if pd.isna(text) or str(text).strip().lower() in stopwords_non_obat:
        return []

    text = str(text).lower()
    text = re.sub(r'\([^)]*\)', '', text)
    text = re.sub(r'[\n→\-]+', ',', text)
    text = re.sub(r'[^a-z0-9, ]', ' ', text)

    tokens = [t.strip() for t in text.split(',') if t.strip()]
    tokens = [canonical_map.get(t, t) for t in tokens]  # mapping typo
    tokens = [t for t in tokens if t in obat_standar]  # filter hanya obat

    return sorted(set(tokens))

# Terapkan
cleaned_data['alergi_obat_clean'] = cleaned_data['alergi_obat'].apply(clean_alergi_obat_strict)

# 5. Gabungkan kembali jadi string konsisten
# =========================
cleaned_data['alergi_obat_str'] = cleaned_data['alergi_obat_clean'].apply(lambda x: ', '.join(x) if x else '')

# =========================
# 6. Buat mapping string unik → angka
# =========================
unique_combinations = cleaned_data['alergi_obat_str'].unique()
mapping_dict = {name: idx for idx, name in enumerate(unique_combinations, start=1)}
mapping_dict[''] = 0  # untuk yang kosong/NaN → 0

# =========================
# 7. Ganti kolom alergi_obat dengan angka
# =========================
cleaned_data['alergi_obat'] = cleaned_data['alergi_obat_str'].map(mapping_dict)

# =========================
# 8. Drop kolom bantu jika mau
# =========================
cleaned_data.drop(columns=['alergi_obat_clean', 'alergi_obat_str'], inplace=True)

# =========================
# 9. Cek hasil
# =========================
print("\nMapping kategori → angka:")
print(mapping_dict)
print(cleaned_data['alergi_obat'].value_counts(dropna=True).head(50))
```

```python
import pandas as pd
import numpy as np
import re
from rapidfuzz import process, fuzz

df = cleaned_data

# === 1. Membersihkan teks ===
def clean_treatment(val):
    if pd.isna(val):
        return np.nan

    val = re.sub(r'\bundefined\b,?\s*', '', val)
    val = str(val).strip().lower()
    val = re.sub(r'\s*\n\s*', ', ', val)
    val = re.sub(r'^\-+\s*', '', val)
    val = re.sub(r'^[,]+\s*', '', val)

    # Hapus pengulangan frasa berturut-turut yang sama
    parts = val.split(', ')
    new_parts = []
    prev = None
    for p in parts:
        if p != prev:
            new_parts.append(p)
        prev = p
    val = ', '.join(new_parts)

    if val == "" or val in {"tidak ada", "none", "n/a"}:
        return np.nan

    # Normalisasi sinonim seperti biasa
    mapping_synonym = {
        'aural toilet': 'aural toilet',
        'nasal toilet': 'nasal toilet',
        'pembersihan hidung': 'pembersihan hidung',
        'periksa tht' : 'pemeriksaan tht',
        'periksta tht' : 'pemeriksaan tht',
        'endoskop[i' : 'endoskopi'
    }
    for key, norm in mapping_synonym.items():
        if key in val:
            val = val.replace(key, norm)

    return val

df['treatment'] = df['treatment'].apply(clean_treatment)

# # === 2. Pisahkan jika ada beberapa tindakan (ambil yang pertama saja) ===
# df['treatment'] = df['treatment'].apply(
#     lambda x: re.split(r'(?<=toilet)|(?<=hidung)\s', x)[0].strip() if pd.notna(x) else np.nan
# )

# === 3. Normalisasi fuzzy matching ===

# Ambil kategori unik hasil pembersihan & pemisahan
unique_treatments = sorted(df['treatment'].dropna().unique())

# Buat daftar kategori standar manual dari unique_treatments (atau definisikan manual sesuai kebutuhan)
# Contoh ambil kategori "benar" dari unique_treatments, kamu bisa ganti ini dengan daftar resmi
standard_categories = standard_categories = [
    'pelayanan poliklinik spesialis, evakuasi/pengambilan cerumen/stosel/corpus alienum ringan non binatang/discharge di telinga hidung, aural toilet',
    'pelayanan poliklinik spesialis, pemeriksaan dokter spesialis pasien baru',
    'pelayanan poliklinik spesialis, evakuasi/pengambilan cerumen/stosel/corpus alienum ringan non binatang/discrge di telinga hidung, nasoendoskopi',
    'pelayanan poliklinik spesialis, pembersihan hidung, nasal toilet',
    'pelayanan poliklinik spesialis, evakuasi/pengambilan cerumen/stosel/corpus alienum ringan non binatang/discrge di telinga hidung',
    'pemeriksaan dokter spesialis pasien lama, provokasi hidung/pengambilan pus dengan vacuum','pemeriksaan tht',
    'pelayanan poliklinik spesialis, endoskopi laring',
    'pelayanan poliklinik spesialis, pembersihan hidung, angkat jahit/hecting aff &gt;10',
    'pelayanan poliklinik spesialis, pembersihan hidung, angkat jahit/hecting aff 6-10',
    'pemeriksaan dokter spesialis pasien lama, pembersihan hidung',
    'pelayanan poliklinik spesialis, pasang/lepas tampon daryantule ditelinga/hidung (10cm), evakuasi/pengambilan cerumen/stosel/corpus alienum ringan non binatang/discrge di telinga hidung, aural toilet',

]

def normalize_fuzzy(cat, standard_list, threshold=85):
    if pd.isna(cat):
        return np.nan
    match = process.extractOne(cat, standard_list, scorer=fuzz.token_sort_ratio)
    if match and match[1] >= threshold:
        return match[0]
    else:
        return cat

# Terapkan normalisasi fuzzy, overwrite kolom treatment langsung
df['treatment'] = df['treatment'].apply(lambda x: normalize_fuzzy(x, standard_categories))

# === 4. Buat mapping kategori → angka ===
unique_treatments = sorted(df['treatment'].dropna().unique())
treatment_to_num = {t: i+1 for i, t in enumerate(unique_treatments)}
treatment_to_num[np.nan] = 0  # kosong → 0

# === 5. Ubah kolom treatment asli menjadi numerik ===
df['treatment'] = df['treatment'].map(treatment_to_num).fillna(0).astype(int)

print("\nMapping kategori ke angka:", treatment_to_num)
print(df['treatment'].value_counts(dropna=True).head(50))
```
